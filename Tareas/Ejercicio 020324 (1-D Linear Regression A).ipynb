Creacin de una funcin que calule automaticamente la linea de regresion automaticamente

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
    #Para cagar info
    data = pd.read_csv("auto-mpg.csv")
```

```python
#Limpiar datos NAN
data.dropna(inplace=True)
```

```python
# Crear el modelo de regresin lineal
model = LinearRegression()
X = data[['horsepower']]
y = data['mpg']
model.fit(X, y)
```

```python
# Predecir los valores
y_pred = model.predict(X)
```

```python
# Calcular el coeficiente de determinacin (R^2)
r2 = model.score(X, y)
```

```python
# Graficar los resultados
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(data['horsepower'], data['mpg'], color='r', linestyle='', marker='o', label='Datos reales')
ax.plot(data['horsepower'], y_pred, linestyle='-', color='b', label='Lnea de regresin')
ax.set_xlabel("Horsepower")
ax.set_ylabel("MPG")
ax.set_title(f"Horsepower vs MPG $R^2$ = {r2}")
ax.legend()
```
